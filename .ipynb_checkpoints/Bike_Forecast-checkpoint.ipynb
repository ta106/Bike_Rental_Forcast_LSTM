```python
def plt_temp(col):
    temp = df[[col,'Total renter','Casual renter' , 'Registered renter']]
    temp[col] = temp[col].apply(lambda r: round(r))
    temp = temp.groupby(col).agg(["count",'sum'])
    headers = set([i[0] for i in temp.columns])
    ## total renters is chosen to filter temps with few recorded hours
    to_plt = temp['Total renter']
    for h in headers:
        line = temp[h]
        to_plt[h + "_normalized"] = line.apply(lambda row: row['sum']/row['count'],axis=1)
    to_plt = to_plt[to_plt['count']> 10]
    
    to_plt.drop(['count','sum'],axis=1).plot()
    return to_plt
```

```python
temp = df
```

```python
timeseries = temp.set_index('datetime').resample('1H').median()
```

```python
timeseries = timeseries.reset_index()
```

```python
## impute weather data with previous hour data
import math
impute_with_lag = ['Temp','Feel-like temp','Humidity','Windspeed']

for j in impute_with_lag:
    res = []
    for i in timeseries[j]:

        if math.isnan(i):
            res.append(res[-1])
        else:
            res.append(i)
    timeseries[j + "_fixed"] = res
```

```python
timeseries['month'] = timeseries['datetime'].apply(lambda r:r.month)
timeseries['hour'] = timeseries['datetime'].apply(lambda r:r.hour)
```

```python
timeseries
```

```python
timeseries['date'] = timeseries['datetime'].apply(lambda r: r.date())
```

```python
daily_info = df.groupby('Date').first().reset_index()[['Date','Weekday','season','Holiday','Workingday','weather']]
```

```python
lookup = {}
for i in daily_info.iterrows():
    lookup[i[1]['Date']] = {
        "Weekday"     :       i[1]['Weekday'],
"season"               :  i[1]['season'],
"Holiday"             :   i[1]['Holiday'],
'Workingday'           : i[1]['Workingday'],
'weather'               : i[1]['weather']

    }
    
```

```python
for i in ['Weekday','season','Holiday','Workingday','weather']:
    timeseries[i+'_fixed'] = timeseries['date'].apply(lambda r: lookup[r][i])
timeseries
```

```python
for i in ['Registered renter','Total renter','Casual renter']:
    timeseries[i] = timeseries[i].fillna(0)
```

```python
timeseries_ = timeseries.drop(['Weekday','season','Holiday','Workingday','Temp','weather','Feel-like temp'\
                               ,
                'Humidity','Windspeed','date'],axis=1)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import Sequence
from datetime import timedelta
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
test = timeseries_[timeseries_['datetime'] >= (timeseries_['datetime'].max() - timedelta(days=60)) ]
```

```python
train = timeseries_[timeseries_['datetime'] < test['datetime'].min()]
```

```python
train.columns
```

```python
train['date'] = train['datetime'].apply(lambda r:r.date())
```

```python
train_ = train[['date','Total renter','Temp_fixed','Humidity_fixed','Windspeed_fixed']].groupby('date').agg({
    'Total renter':'sum','Temp_fixed':"mean",'Humidity_fixed':'mean','Windspeed_fixed':'mean'
})
```

```python
train_matrix = train[['Total renter','Temp_fixed','Humidity_fixed','Windspeed_fixed']]
```

```python
train_matrix = train_matrix.values
```

```python
train_matrix
```

```python
train_matrix.shape
```

```python
sc = MinMaxScaler()
training_set_scaled = sc.fit_transform(train_matrix)
```

```python
X_train = []
y_train = []

n_past = 168 *2     # Number of past hours we want to use to predict the future

for i in range(n_past, len(training_set_scaled)  ):
    X_train.append(training_set_scaled[i - n_past:i, 0:train_matrix.shape[1] ])
    y_train.append(training_set_scaled[i   , 0])

X_train, y_train = np.array(X_train), np.array(y_train)

print('X_train shape == {}.'.format(X_train.shape))
print('y_train shape == {}.'.format(y_train.shape))
```

```python

# Import Libraries and packages from Keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
# Initializing the Neural Network based on LSTM
model = Sequential()

# Adding 1st LSTM layer
model.add(LSTM(units=200, return_sequences=True, input_shape=(n_past, train_matrix.shape[1])))

# Adding 2nd LSTM layer
model.add(LSTM(units=10, return_sequences=False))

# Output layer
model.add(Dense(units=1, activation='linear'))

# Compiling the Neural Network
model.compile(optimizer ='adam' , loss='mean_squared_error')
```

```python
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, TensorBoard
import time
es = EarlyStopping(monitor='val_loss', min_delta=1e-10, patience=10, verbose=1)
rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, verbose=1)
mcp = ModelCheckpoint(filepath='weights.h5', monitor='val_loss', verbose=1, save_best_only=True, save_weights_only=True)

tb = TensorBoard('logs')

history = model.fit(X_train, y_train, shuffle=True, epochs=30, callbacks=[es, rlr, mcp, tb], validation_split=0.2, verbose=1, batch_size=256)
```
